## 6. Histograms

Histograms are a type of bar plot that shows the frequency or number of data points that fall within specified ranges (bins). They are useful for understanding the distribution of numerical data.

### 6.1 Understanding Histograms

- **Bins:** Intervals that divide the entire range of data into a series of segments. Each bin represents the frequency of data within that range.
- **Frequency:** The number of data points that fall within each bin.

### 6.2 Creating a Simple Histogram

Let's create a histogram of randomly generated data to see how it works.

**Example: Histogram of Normally Distributed Data**



### **Explanation:**

- `np.random.randn(1000)` generates 1000 random numbers from a standard normal distribution (mean = 0, standard deviation = 1).
- `plt.hist(data)` creates the histogram.

---

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.randn(1000)

# Create histogram
plt.hist(data)

# Add title and labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
````

### 6.3 Customizing Histograms

You can customize histograms to better represent your data.

#### **Common Customizations**

- **Bins:** Change the number of bins to adjust the granularity.
- **Color and Edge Color:** Customize the color of the bars and their edges.
- **Density:** Normalize the histogram so that the area under the histogram integrates to 1 (useful for comparing with probability density functions).

**Example: Histogram with More Bins and Custom Colors**

```python

plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Add title and labels

```

### **Overlaying a Density Curve**

You can overlay a probability density function (PDF) on the histogram to compare the empirical data distribution with a theoretical one.

## 7. Pie Charts

Pie charts are circular charts divided into sectors, illustrating numerical proportions in a dataset.

### 7.1 Understanding Pie Charts

- **Slices:** Each slice represents a category's contribution to the whole.
- **Angles:** The angle of each slice is proportional to the frequency or percentage it represents.

### 7.2 Creating a Simple Pie Chart

Let's create a pie chart showing market share of different companies.

**Example: Market Share Pie Chart**



### **Explanation:**

- `plt.pie()` creates the pie chart.
- `autopct='%1.1f%%'` displays the percentage value on each slice.
- `startangle=140` rotates the chart for better visualization.
- `plt.axis('equal')` ensures the pie chart is a circle.

---

````python
# Data
companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [40, 25, 20, 15]

# Create pie chart
plt.pie(market_share, labels=companies, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add title
plt.title('Market Share of Companies')

# Show plot
plt.show()
````

## 8. Subplots

Subplots allow you to display multiple plots in a single figure, enabling comparisons and more complex data visualization layouts.

### 8.1 Creating Multiple Plots in One Figure

Let's create a figure with two subplots: a sine wave and a cosine wave.

**Example: Sine and Cosine Waves in Subplots**



### **Explanation:**

- `plt.subplots(nrows=1, ncols=2)` creates a figure with 1 row and 2 columns of subplots.
- `axes[0]` and `axes[1]` are the two subplot axes.

---

````python
# Create data
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# First subplot
axes[0].plot(x, y_sin, color='blue')
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('Angle [radians]')
axes[0].set_ylabel('sin(x)')

# Second subplot
axes[1].plot(x, y_cos, color='red')
axes[1].set_title('Cosine Wave')
axes[1].set_xlabel('Angle [radians]')
axes[1].set_ylabel('cos(x)')

#Adjust the spacing between subplots and elements in a figure to ensure that everything fits properly within the figure area without overlap.
plt.tight_layout()

# Show plots

plt.show()
````
